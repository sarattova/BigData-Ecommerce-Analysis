```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import col, to_timestamp, hour, round, sum, desc, when
from pyspark.sql.window import Window
import matplotlib.pyplot as plt
import seaborn as sns

# Инициализируем управляющий объект - SparkSession
spark = SparkSession.builder \
    .appName("Ecommerce_BigData_Project") \
    .getOrCreate()

# Читаем CSV-файл. header=True учитывает имена колонок, inferSchema=True автоматически определяет типы данных
df = spark.read.csv("ecommerce_data.csv", header=True, inferSchema=True)

# Отображаем структуру (схему) данных для проверки корректности считывания
print("Структура исходного датасета:")
df.printSchema()

# Выводим первые 5 строк таблицы для наглядности
df.show(5)
```

```python
# 1. Удаляем строки, где отсутствуют критически важные бизнес-идентификаторы
df_clean = df.na.drop(subset=["CustomerID", "Description"])

# 2. Фильтруем аномалии: исключаем отрицательные или нулевые объемы и цены транзакций
df_clean = df_clean.filter((col("Quantity") > 0) & (col("UnitPrice") > 0))

# 3. Переводим строковый формат даты в полноценный объект Timestamp для временного анализа
df_clean = df_clean.withColumn("InvoiceDate", to_timestamp(col("InvoiceDate"), "M/d/yyyy H:mm"))

# 4. Вычисляем общую сумму для каждой строки (Количество товара * Цена за единицу)
df_clean = df_clean.withColumn("TotalAmount", round(col("Quantity") * col("UnitPrice"), 2))

# Сравниваем объемы данных до и после очистки логов
print(f"Количество строк до фильтрации: {df.count()}")
print(f"Количество строк после удаления аномалий: {df_clean.count()}")
df_clean.show(5)
```

```python
# А) Географический анализ: группировка выручки по странам (выделяем Топ-5)
geo_analysis = df_clean.groupBy("Country") \
    .agg(round(sum("TotalAmount"), 2).alias("CountryRevenue")) \
    .sort(desc("CountryRevenue"))

print("--- ТОП-5 СТРАН ПО ВЫРУЧКЕ ---")
geo_analysis.show(5)

# Б) Временной анализ: выделяем час транзакции и считаем количество операций для оценки нагрузки
df_with_hour = df_clean.withColumn("InvoiceHour", hour(col("InvoiceDate")))
hourly_load = df_with_hour.groupBy("InvoiceHour").count().sort("InvoiceHour")

print("--- РАСПРЕДЕЛЕНИЕ НАГРУЗКИ ПО ЧАСАМ СУТОК ---")
hourly_load.show(24)
```

```python
# 1. Агрегируем суммарную выручку по каждому уникальному товару
product_revenue = df_clean.groupBy("StockCode", "Description").agg(sum("TotalAmount").alias("ProdRevenue"))

# 2. Извлекаем общую глобальную выручку всего магазина в локальную переменную
total_shop_revenue = product_revenue.select(sum("ProdRevenue")).collect()[0][0]

# 3. Создаем спецификацию окна для ранжирования товаров по убыванию их прибыльности
window_spec = Window.orderBy(desc("ProdRevenue"))

# Вычисляем долю каждого товара и кумулятивный (накопительный) процент с помощью окна
product_cum_percentage = product_revenue \
    .withColumn("RevenueShare", col("ProdRevenue") / total_shop_revenue) \
    .withColumn("CumPercentage", sum("RevenueShare").over(window_spec) * 100)

# 4. Классифицируем товары: класс A (первые 80% выручки), класс B (следующие 15%), класс C (оставшиеся 5%)
abc_result = product_cum_percentage.withColumn(
    "ABC_Class",
    when(col("CumPercentage") <= 80, "A")
    .when((col("CumPercentage") > 80) & (col("CumPercentage") <= 95), "B")
    .otherwise("C")
)

print("--- РЕЗУЛЬТАТЫ РАСПРЕДЕЛЕННОГО ABC-АНАЛИЗА ---")
abc_result.groupBy("ABC_Class").count().show()
```

```python
# Конвертируем агрегированные DataFrame Spark в Pandas (это безопасно для памяти, так как строк мало)
hourly_pd = hourly_load.toPandas()
geo_pd = geo_analysis.limit(5).toPandas()

# Устанавливаем стиль графиков
sns.set_theme(style="whitegrid")

# График 1: Временная нагрузка на систему
plt.figure(figsize=(12, 5))
sns.lineplot(data=hourly_pd, x="InvoiceHour", y="count", marker="o", color="#1f77b4", linewidth=2.5)
plt.title("Транзакционная активность покупателей по часам суток", fontsize=14, fontweight='bold')
plt.xlabel("Час суток (0 - 23)", fontsize=11)
plt.ylabel("Количество обработанных заказов", fontsize=11)
plt.xticks(range(0, 24))
plt.grid(True, linestyle="--", alpha=0.6)
plt.show()

# График 2: Географическая структура продаж
plt.figure(figsize=(10, 4))
sns.barplot(data=geo_pd, x="CountryRevenue", y="Country", hue="Country", palette="Blues_r", legend=False)
plt.title("Топ-5 стран по объемам сгенерированной выручки", fontsize=14, fontweight='bold')
plt.xlabel("Сумма продаж", fontsize=11)
plt.ylabel("Страна", fontsize=11)
plt.tight_layout()
plt.show()
```
